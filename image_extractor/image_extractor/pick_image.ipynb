```python
import os
import cv2

def extract_every_third_image(input_folder, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    image_files = [f for f in os.listdir(input_folder) if f.endswith(('.png', '.jpg', '.jpeg'))]
    image_files.sort()  # Sort the filenames to maintain their original order
    
    for idx, image_file in enumerate(image_files):
        if idx % 3 == 0:  # Extract every third image
            image_path = os.path.join(input_folder, image_file)
            output_path = os.path.join(output_folder, f'image_{idx:04d}.png')
            
            img = cv2.imread(image_path)
            cv2.imwrite(output_path, img)
            print(f'Saved image: {output_path}')
```

```python
# Set the input and output folder paths
input_folder = '/home/yh/datasets/from_me/extracted_images'
output_folder = '/home/yh/datasets/from_me/sorted_images'

# Call the function to extract and save the images
extract_every_third_image(input_folder, output_folder)
```
